# 70. Climbing Stairs  
**Category:** Dynamic Programming / Counting  
**Difficulty:** Easy  

You are climbing a staircase.  
It takes **n steps** to reach the top.

Each time you can climb **1 or 2 steps**.

### Goal  
Return the **number of distinct ways** to climb to the top.

### Example  
**Input:** `n = 3`  
**Output:** `3`  
**Explanation:**  
1+1+1  
1+2  
2+1  


# 🎯 Intuition & ML Relevance

This problem is a classic **dynamic programming counting problem**, directly related to:

- Sequence modeling  
- Recurrence relations (like RNNs)  
- Fibonacci‑style growth patterns  
- Probabilistic modeling  
- Path counting in graphs  
- Reinforcement learning (ways to reach a state)

The key idea:

> To reach step `i`, you must come from either step `i-1` (1 step) or step `i-2` (2 steps).

This gives the recurrence:



\[
dp[i] = dp[i-1] + dp[i-2]
\]



This is identical to the Fibonacci sequence.


# 🟦 Optimized DP Solution

```python
def climbStairs(n):
    if n <= 2:
        return n

    prev2 = 1   # ways to reach step 1
    prev1 = 2   # ways to reach step 2

    for _ in range(3, n + 1):
        current = prev1 + prev2
        prev2 = prev1
        prev1 = current

    return prev1

# Example
climbStairs(3)
```

# ⚙️ Dynamic Programming Explanation

To reach step `i`, you have two choices:

- From step `i-1` (take 1 step)
- From step `i-2` (take 2 steps)

So the recurrence is:



\[
dp[i] = dp[i-1] + dp[i-2]
\]



This is exactly the Fibonacci sequence.

We optimize space by storing only two variables:

- `prev1` → ways to reach step `i-1`  
- `prev2` → ways to reach step `i-2`  


# ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
We compute each step once.

### Space Complexity: **O(1)**  
We only store two variables:

- `prev1`  
- `prev2`  

This is optimal and extremely efficient.


# 🟦 DP Array Version (More Explicit)

```python
def climbStairs_dp(n):
    if n <= 2:
        return n

    dp = [0] * (n + 1)
    dp[1] = 1
    dp[2] = 2

    for i in range(3, n + 1):
        dp[i] = dp[i-1] + dp[i-2]

    return dp[n]

# Example
climbStairs_dp(3)
```

# ⚠️ Edge Cases

- `n = 1` → only 1 way  
- `n = 2` → two ways (1+1, 2)  
- Large `n` (performance still excellent)  
- No negative inputs (problem guarantees n ≥ 1)
