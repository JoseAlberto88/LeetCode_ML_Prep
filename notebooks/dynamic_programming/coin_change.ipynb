# 322. Coin Change  
**Category:** Dynamic Programming / Optimization  
**Difficulty:** Medium  

You are given an integer array `coins` representing coin denominations  
and an integer `amount` representing a target value.

### Goal  
Return the **fewest number of coins** needed to make up the amount.  
If it is not possible, return **-1**.

### Example  
**Input:** `coins = [1,2,5]`, `amount = 11`  
**Output:** `3`  
**Explanation:** 11 = 5 + 5 + 1  


# 🎯 Intuition & ML Relevance

This problem is a classic **dynamic programming optimization** task.

It connects to ML concepts such as:

- Cost minimization  
- Optimal resource allocation  
- Dynamic programming over discrete states  
- Reinforcement learning (minimum steps to reach a state)  
- Graph shortest-path analogies (each amount is a node)

The key idea:

> To compute the minimum coins for `amount`,  
> try every coin and choose the best option.

This leads to the recurrence:



\[
dp[x] = \min(dp[x],\ dp[x - coin] + 1)
\]



This is a bottom‑up DP solution.


# 🟦 Bottom-Up DP Solution

```python
def coinChange(coins, amount):
    # dp[x] = minimum coins needed for amount x
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # base case

    for x in range(1, amount + 1):
        for coin in coins:
            if x - coin >= 0:
                dp[x] = min(dp[x], dp[x - coin] + 1)

    return dp[amount] if dp[amount] != float('inf') else -1

# Example
coinChange([1,2,5], 11)
```

# ⚙️ Dynamic Programming Explanation

We build a DP array where:

- `dp[x]` = minimum coins needed to make amount `x`
- `dp[0] = 0` (zero coins needed for amount zero)

For each amount `x` from 1 to `amount`:

Try every coin:

- If `coin` fits (x - coin ≥ 0),  
  then we can form `x` by using:
  


\[
dp[x - coin] + 1
\]



We take the minimum over all choices.

This is similar to:

- Shortest path in a DAG  
- Minimum cost to reach a state  
- RL value iteration  


# ⏱️ Time & Space Complexity

### Time Complexity: **O(amount × number_of_coins)**  
We compute each amount using all coins.

### Space Complexity: **O(amount)**  
We store a DP array of size `amount + 1`.


# 🟦 Early Exit BFS Version (Optional)

```python
from collections import deque

def coinChange_bfs(coins, amount):
    if amount == 0:
        return 0

    queue = deque([(0, 0)])  # (current_amount, steps)
    visited = set([0])

    while queue:
        curr, steps = queue.popleft()

        for coin in coins:
            nxt = curr + coin
            if nxt == amount:
                return steps + 1
            if nxt < amount and nxt not in visited:
                visited.add(nxt)
                queue.append((nxt, steps + 1))

    return -1
```

# ⚠️ Edge Cases

- `amount = 0` → return 0  
- No combination possible → return -1  
- Large amount values (DP still efficient)  
- Duplicate coin values (safe to ignore)
