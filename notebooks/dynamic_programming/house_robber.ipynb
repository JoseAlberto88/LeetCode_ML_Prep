# 198. House Robber  
**Category:** Dynamic Programming / Optimization  
**Difficulty:** Medium  

You are a professional robber planning to rob houses along a street.  
Each house has a certain amount of money, but **you cannot rob two adjacent houses**.

### Example
**Input:** `nums = [1,2,3,1]`  
**Output:** `4`  
**Explanation:** Rob house 1 (1) + house 3 (3) → total = 4.


# 🎯 Intuition & ML Relevance

This problem is a classic example of **dynamic programming with constraints**, which maps directly to ML concepts such as:

- Sequential decision-making  
- Reinforcement learning reward maximization  
- Feature extraction from time-series with adjacency constraints  
- Optimization under structural restrictions  
- State transitions with memory (RNN-like reasoning)

The key idea:

> At each house, choose between **robbing it** (and skipping the previous one) or **skipping it** (and keeping the best so far).

This demonstrates:

- Local vs global optimality  
- State recurrence  
- Efficient O(n) optimization  
- Avoiding overlapping subproblems  


# 🟦 Optimized DP Solution

```python
def rob(nums):
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0]

    prev2 = nums[0]      # Max money up to house i-2
    prev1 = max(nums[0], nums[1])  # Max money up to house i-1

    for i in range(2, len(nums)):
        # Either rob this house + prev2, or skip it and keep prev1
        current = max(prev1, prev2 + nums[i])
        prev2 = prev1
        prev1 = current

    return prev1

# Example
rob([1,2,3,1])
```

# ⚙️ Dynamic Programming Explanation

We track two states:

### 1. `prev1`
Best amount we can rob up to the previous house.

### 2. `prev2`
Best amount we can rob up to the house before the previous one.

At each house `i`, we compute:



\[
dp[i] = \max(dp[i-1],\ dp[i-2] + nums[i])
\]



Meaning:

- **Skip house i** → keep `dp[i-1]`  
- **Rob house i** → take `nums[i] + dp[i-2]`  

This ensures we never rob adjacent houses.


# ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
We scan the list once.

### Space Complexity: **O(1)**  
We only store two variables:

- `prev1`  
- `prev2`  

This is optimal and extremely efficient.


# 🟦 Alternative DP Version

```python
def rob_dp(nums):
    if not nums:
        return 0
    if len(nums) == 1:
        return nums[0]

    dp = [0] * len(nums)
    dp[0] = nums[0]
    dp[1] = max(nums[0], nums[1])

    for i in range(2, len(nums)):
        dp[i] = max(dp[i-1], dp[i-2] + nums[i])

    return dp[-1]

# Example
rob_dp([1,2,3,1])
```

# ⚠️ Edge Cases

- Empty list → return `0`  
- Single house → return its value  
- All houses have the same value  
- Large arrays (performance)  
- Very large values (Python handles big ints safely)
