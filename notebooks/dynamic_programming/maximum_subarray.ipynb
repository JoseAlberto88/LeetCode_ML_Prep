# 53. Maximum Subarray  
**Category:** Dynamic Programming / Optimization  
**Difficulty:** Medium  

Given an integer array `nums`, find the **contiguous subarray** with the largest sum and return that sum.

Example:  
Input: nums = [-2,1,-3,4,-1,2,1,-5,4]  
Output: 6  
Explanation: The subarray [4,-1,2,1] has the largest sum = 6.


## 🎯 Intuition & ML Relevance

This problem is a classic example of **dynamic programming** and **optimization**, which are fundamental in:

- Loss minimization  
- Gradient-based optimization  
- Time-series feature extraction  
- Signal smoothing  
- Reinforcement learning (max reward segments)  
- Portfolio optimization  

Kadane’s Algorithm is a perfect demonstration of:
- Local vs global optimality  
- State transitions  
- Efficient O(n) optimization  


```python
def maxSubArray(nums):
    current_sum = nums[0]
    max_sum = nums[0]

    for num in nums[1:]:
        # Either extend the current subarray or start a new one
        current_sum = max(num, current_sum + num)
        max_sum = max(max_sum, current_sum)

    return max_sum

# Example
maxSubArray([-2,1,-3,4,-1,2,1,-5,4])
```

## ⚙️ Kadane’s Algorithm Explanation

We track two values:

1. **current_sum**  
   - Best subarray ending at the current index  
   - Either:
     - Extend previous subarray (`current_sum + num`)
     - Start new subarray (`num`)

2. **max_sum**  
   - Best subarray seen so far

This is a classic DP recurrence:



\[
current\_sum[i] = \max(nums[i], current\_sum[i-1] + nums[i])
\]





\[
max\_sum = \max(max\_sum, current\_sum[i])
\]



This gives an optimal solution in **O(n)** time.


## ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**
We scan the array once.

### Space Complexity: **O(1)**
We only store two variables:
- current_sum  
- max_sum  

This is optimal and extremely efficient.


```python
def maxSubArray_dp(nums):
    dp = [0] * len(nums)
    dp[0] = nums[0]
    max_sum = nums[0]

    for i in range(1, len(nums)):
        dp[i] = max(nums[i], dp[i-1] + nums[i])
        max_sum = max(max_sum, dp[i])

    return max_sum

# Example
maxSubArray_dp([-2,1,-3,4,-1,2,1,-5,4])
```

## ⚠️ Edge Cases

- Single element array → return that element  
- All negative numbers → return the largest (least negative)  
- Very large arrays (performance test)  
- Mixed positive/negative sequences
