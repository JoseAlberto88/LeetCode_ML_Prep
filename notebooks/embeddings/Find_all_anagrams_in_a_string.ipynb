# 4. Find All Anagrams in a String  
**Category:** Embeddings / Similarity / Sliding Window  
**Difficulty:** Medium  

Given two strings `s` and `p`, return all starting indices in `s` where an anagram of `p` appears.

Example:  
s = "cbaebabacd", p = "abc"  
Output: [0, 6]


## 🎯 Intuition & ML Relevance

This problem is fundamentally about **detecting similarity between two sequences** using lightweight vector representations — a core idea in many ML systems.

Why this matters:

- Each substring window in `s` is represented as a **frequency embedding** (a 26‑dimensional vector).
- The pattern `p` is also represented as a frequency embedding.
- An anagram match occurs when the two embeddings are **identical** — similar to comparing feature vectors.

This mirrors ML concepts such as:
- Bag‑of‑words embeddings  
- Token frequency vectors  
- Sliding‑window feature extraction  
- Local similarity search  
- Sparse vector updates  

This problem teaches:
- Efficient embedding comparison  
- Window‑based pattern detection  
- Incremental state updates  
- Hash‑based feature tracking  


## 🧩 Code Implementation (Sliding Window + Embeddings)

```python
from collections import Counter

def findAnagrams(s: str, p: str):
    if len(p) > len(s):
        return []

    result = []
    p_count = Counter(p)
    window = Counter(s[:len(p)])

    # Check the first window
    if window == p_count:
        result.append(0)

    left = 0
    for right in range(len(p), len(s)):
        # Add the new character on the right
        window[s[right]] += 1

        # Remove the character on the left
        window[s[left]] -= 1
        if window[s[left]] == 0:
            del window[s[left]]
        left += 1

        # Compare embeddings
        if window == p_count:
            result.append(left)

    return result

# Example
findAnagrams("cbaebabacd", "abc")
```

## 🪟 Sliding Window Logic

An anagram match occurs when the **frequency embedding** of the current window equals the embedding of `p`.

Sliding window steps:
1. Build the frequency embedding for `p`.  
2. Build the embedding for the first window in `s`.  
3. Slide the window across `s`:  
   - Add the new character  
   - Remove the old character  
4. Compare embeddings at each step.  
5. If equal → record the index.

Why this works:
- Only two characters change per step  
- No full recomputation  
- Embedding comparison is constant time  

This is the same principle used in:
- Rolling hash algorithms  
- Convolutional filters  
- Attention windows  


## ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
Each character is added and removed at most once.

### Space Complexity: **O(1)**  
The alphabet is fixed (26 lowercase letters), so the embedding size is constant.

Efficient and scalable — ideal for large‑scale text preprocessing.


## 🧮 Array-Based Optimization (Faster Than Counter)

```python
def findAnagrams_array(s: str, p: str):
    if len(p) > len(s):
        return []

    result = []
    p_count = [0] * 26
    window = [0] * 26

    for ch in p:
        p_count[ord(ch) - ord('a')] += 1

    for ch in s[:len(p)]:
        window[ord(ch) - ord('a')] += 1

    if window == p_count:
        result.append(0)

    left = 0
    for right in range(len(p), len(s)):
        window[ord(s[right]) - ord('a')] += 1
        window[ord(s[left]) - ord('a')] -= 1
        left += 1

        if window == p_count:
            result.append(left)

    return result
```

## ⚠️ Edge Cases

- `p` longer than `s` → no matches  
- No anagrams found  
- All characters identical  
- Overlapping anagrams  
- Very large strings (performance test)
