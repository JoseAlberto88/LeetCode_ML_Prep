# 5. Group Anagrams  
**Category:** Embeddings / Hashing / Similarity  
**Difficulty:** Medium  

Given an array of strings, group the anagrams together.

Example:  
Input: ["eat", "tea", "tan", "ate", "nat", "bat"]  
Output: [["eat","tea","ate"], ["tan","nat"], ["bat"]]


## 🎯 Intuition & ML Relevance

This problem is about grouping items that share the **same embedding representation**.

Why this matters:

- Each word can be represented as a **frequency embedding** (26‑dimensional vector).
- Anagrams share the **same embedding**, even if their order differs.
- Grouping anagrams is equivalent to grouping identical embeddings.

This mirrors ML concepts such as:
- Token frequency vectors  
- Bag‑of‑words embeddings  
- Hashing for similarity detection  
- Clustering identical feature vectors  
- Canonical representation of inputs  

This problem teaches:
- How to build stable embeddings  
- How to use hashing for grouping  
- How to design canonical forms for comparison  


## 🧩 Code Implementation (Hash Map + Embeddings)

```python
from collections import defaultdict

def groupAnagrams(strs):
    groups = defaultdict(list)

    for word in strs:
        # Build a 26-dimensional frequency embedding
        freq = [0] * 26
        for ch in word:
            freq[ord(ch) - ord('a')] += 1
        
        # Use the embedding as a hashable key
        groups[tuple(freq)].append(word)

    return list(groups.values())

# Example
groupAnagrams(["eat", "tea", "tan", "ate", "nat", "bat"])
```

## 🧠 Embedding Logic

Two words are anagrams if:
- They contain the same characters  
- With the same frequencies  
- Regardless of order  

So each word becomes a **frequency embedding**:
- "eat" → [1,0,0,0,1,0,...]  
- "tea" → [1,0,0,0,1,0,...]  
- "ate" → [1,0,0,0,1,0,...]  

All identical → same group.

This is identical to:
- Comparing sparse vectors  
- Canonicalizing inputs  
- Hashing embeddings for grouping  


## 🪟 Alternative Sorted-Key Approach

```python
def groupAnagrams_sorted(strs):
    groups = defaultdict(list)

    for word in strs:
        key = ''.join(sorted(word))  # canonical form
        groups[key].append(word)

    return list(groups.values())
```

## ⏱️ Time & Space Complexity

### Frequency-Embedding Method
- Time: **O(n · k)**  
  (n = number of words, k = max word length)
- Space: **O(n · k)** for storing groups

### Sorted-Key Method
- Time: **O(n · k log k)**  
- Space: **O(n · k)**

The embedding method is faster for long strings.


## ⚠️ Edge Cases

- Empty list → return []  
- Single word → its own group  
- Words with different lengths  
- Large input sizes (performance)  
- Mixed-case or unicode (if extended)
