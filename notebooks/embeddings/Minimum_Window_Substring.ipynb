# 6. Minimum Window Substring  
**Category:** Sliding Window / Hash Maps / Substring Search  
**Difficulty:** Hard  

Given two strings `s` and `t`, return the **minimum window substring** of `s` such that every character in `t` appears in the window (including duplicates).

If no such substring exists, return an empty string.

Example:  
s = "ADOBECODEBANC", t = "ABC"  
Output: "BANC"


## 🎯 Intuition & ML Relevance

This problem is about **finding the smallest substring that satisfies a constraint**, which is deeply related to:

- Attention windowing  
- Context selection in NLP  
- Dynamic feature extraction  
- Constraint‑based sequence scanning  
- Efficient state tracking  

Why this matters for ML:

- We maintain **frequency embeddings** for the target string `t`.
- We slide a window over `s` and maintain a **dynamic embedding** of the current window.
- The goal is to find the *smallest* window whose embedding **covers** the embedding of `t`.

This mirrors:
- Attention masks  
- Query‑key matching  
- Coverage constraints in sequence models  
- Efficient scanning of long sequences  

This problem teaches:
- How to maintain counts efficiently  
- How to shrink and expand windows optimally  
- How to track when constraints are satisfied  


## 🧩 Code Implementation (Optimized Sliding Window)

```python
from collections import Counter

def minWindow(s: str, t: str) -> str:
    if not t or not s:
        return ""

    t_count = Counter(t)
    required = len(t_count)

    left = 0
    formed = 0
    window_count = {}

    ans = (float("inf"), 0, 0)  # window length, left, right

    for right, char in enumerate(s):
        window_count[char] = window_count.get(char, 0) + 1

        if char in t_count and window_count[char] == t_count[char]:
            formed += 1

        # Try to shrink the window
        while left <= right and formed == required:
            if right - left + 1 < ans[0]:
                ans = (right - left + 1, left, right)

            # Remove leftmost character
            left_char = s[left]
            window_count[left_char] -= 1
            if left_char in t_count and window_count[left_char] < t_count[left_char]:
                formed -= 1

            left += 1

    return "" if ans[0] == float("inf") else s[ans[1]:ans[2] + 1]

# Example
minWindow("ADOBECODEBANC", "ABC")
```

## 🪟 Sliding Window Logic

We maintain two pointers:  
- `right` → expands the window  
- `left` → shrinks the window  

Steps:
1. Expand `right` until the window contains all characters of `t`.  
2. Once valid, shrink from the left to find the **smallest** valid window.  
3. Track the best (minimum) window seen so far.  
4. Continue until the entire string is processed.

Why this works:
- Expanding ensures we eventually satisfy the constraint.  
- Shrinking ensures we find the *minimum* window.  
- The window only moves forward → **O(n)** time. 

## ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
Each character is visited at most twice (once by `right`, once by `left`).

### Space Complexity: **O(1)**  
Character counts use fixed-size hash maps (bounded by ASCII/Unicode set).


## ⚠️ Edge Cases

- `t` longer than `s` → no window  
- No valid window exists  
- Repeated characters in `t`  
- Case sensitivity matters  
- Large strings (performance test)
