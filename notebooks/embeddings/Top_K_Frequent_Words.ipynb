# 7. Top K Frequent Words  
**Category:** Embeddings / Frequency Analysis / Priority Queues  
**Difficulty:** Medium  

Given an array of strings `words` and an integer `k`, return the **k most frequent words**.

The output must be sorted by:
1. Highest frequency  
2. Alphabetical order (if frequencies tie)

Example:  
Input: words = ["i","love","leetcode","i","love","coding"], k = 2  
Output: ["i", "love"]


## 🎯 Intuition & ML Relevance

This problem is about ranking items based on **frequency embeddings**, which is central to many ML and NLP systems.

Why this matters:

- Each word is represented by a **frequency count** — a simple but powerful embedding.
- Ranking based on frequency is similar to:
  - TF-IDF weighting  
  - Token importance scoring  
  - Vocabulary pruning  
  - Feature selection in NLP  

This mirrors ML tasks such as:
- Identifying important tokens  
- Ranking features by significance  
- Building word distributions  
- Constructing bag-of-words models  

This problem teaches:
- How to build frequency embeddings  
- How to sort by multiple criteria  
- How to use heaps for efficient top‑k retrieval  


## 🧩 Code Implementation (Heap + Sorting)

```python
from collections import Counter
import heapq

def topKFrequent(words, k):
    freq = Counter(words)

    # Min-heap with custom ordering:
    # (-frequency, word) ensures:
    #   - higher frequency comes first
    #   - alphabetical order breaks ties
    heap = [(-count, word) for word, count in freq.items()]
    heapq.heapify(heap)

    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap)[1])

    return result

# Example
topKFrequent(["i","love","leetcode","i","love","coding"], 2)
```

## 🧠 Frequency Embedding Logic

Each word is mapped to a **frequency embedding**:
- "i" → 2  
- "love" → 2  
- "leetcode" → 1  
- "coding" → 1  

To extract the top‑k:
- Use a heap to efficiently retrieve the highest-frequency items.
- Use alphabetical order to break ties.

This is identical to:
- Ranking tokens by importance  
- Selecting top features  
- Prioritizing high-signal items in NLP pipelines  


## 🪟 Alternative Sorting Approach

```python
def topKFrequent_sort(words, k):
    freq = Counter(words)

    # Sort by:
    # 1. frequency descending
    # 2. alphabetical ascending
    sorted_words = sorted(freq.keys(), key=lambda w: (-freq[w], w))

    return sorted_words[:k]
```

## ⏱️ Time & Space Complexity

### Heap Method
- Time: **O(n log n)**  
- Space: **O(n)**  

### Sorting Method
- Time: **O(n log n)**  
- Space: **O(n)**  

Both are efficient for large vocabularies.


## ⚠️ Edge Cases

- k = 1  
- All words identical  
- All words unique  
- Ties in frequency (alphabetical order matters)  
- Large vocabulary sizes
