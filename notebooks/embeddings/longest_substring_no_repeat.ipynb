# 3. Longest Substring Without Repeating Characters  
**Category:** Embeddings / Similarity / Sliding Window  
**Difficulty:** Medium  

This problem asks for the length of the longest substring without repeating characters.  
Example:  
Input: "abcabcbb" → Output: 3 ("abc")

```python
def lengthOfLongestSubstring(s: str) -> int:
    char_set = set()
    left = 0
    max_len = 0

    for right in range(len(s)):
        # If character repeats, shrink window from the left
        while s[right] in char_set:
            char_set.remove(s[left])
            left += 1
        
        # Expand window
        char_set.add(s[right])
        max_len = max(max_len, right - left + 1)

    return max_len

# Example
lengthOfLongestSubstring("abcabcbb")
```

## 🪟 Sliding Window Explanation

We maintain a window `[left, right]` that always contains **unique characters**.

Steps:
1. Expand `right` pointer through the string.
2. If a character repeats:
   - Shrink the window from the left until the duplicate is removed.
3. Track the maximum window size.

This ensures:
- Each character is added/removed at most once.
- The algorithm runs in **O(n)** time.


## ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**
Each character is visited at most twice (once by `right`, once by `left`).

### Space Complexity: **O(min(n, k))**
Where k = size of the character set (ASCII → 128, Unicode → larger).

Efficient and scalable — ideal for ML preprocessing pipelines.


```python
def lengthOfLongestSubstring_index(s: str) -> int:
    index_map = {}
    left = 0
    max_len = 0

    for right, char in enumerate(s):
        if char in index_map and index_map[char] >= left:
            left = index_map[char] + 1
        
        index_map[char] = right
        max_len = max(max_len, right - left + 1)

    return max_len

# Example
lengthOfLongestSubstring_index("pwwkew")
```

## ⚠️ Edge Cases

- Empty string → return 0  
- All unique characters → return len(s)  
- All same characters → return 1  
- Unicode characters (e.g., emojis)  
- Very long strings (performance test)
