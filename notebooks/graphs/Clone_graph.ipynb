# 133. Clone Graph  
**Category:** Graphs / DFS / BFS / Hash Maps  
**Difficulty:** Medium  

Given a reference to a node in a **connected undirected graph**, return a **deep copy** (clone) of the graph.

Each node contains:
- `val` — unique integer  
- `neighbors` — list of adjacent nodes  

Example graph:
1 -- 2  
|    |  
4 -- 3  

Output: A new graph with identical structure but **new node objects**.


## 🎯 Intuition & ML Relevance

This problem is about **graph traversal + mapping original nodes to cloned nodes**, which is essential in:

- Graph neural networks  
- Knowledge graph duplication  
- Scene graph transformations  
- Copying computational graphs  
- Cloning stateful structures in ML pipelines  

Why this matters:

- Each node must be copied **once**.  
- All edges must be recreated.  
- A mapping from original → clone is required to avoid infinite loops.  

This mirrors ML concepts such as:
- Node embedding propagation  
- Graph reconstruction  
- Message passing networks  
- Maintaining identity mappings across graph copies  


## 🧩 Code Implementation (DFS)

```python
class Node:
    def __init__(self, val=0, neighbors=None):
        self.val = val
        self.neighbors = neighbors if neighbors is not None else []

def cloneGraph(node):
    if not node:
        return None

    old_to_new = {}

    def dfs(curr):
        if curr in old_to_new:
            return old_to_new[curr]

        # Clone the node
        copy = Node(curr.val)
        old_to_new[curr] = copy

        # Clone neighbors
        for nei in curr.neighbors:
            copy.neighbors.append(dfs(nei))

        return copy

    return dfs(node)
```

## 🧠 Graph Cloning Logic

To clone a graph, we must:

1. Create a **new node** for each original node.  
2. Store a mapping:  
   original_node → cloned_node  
3. Recursively clone all neighbors.  
4. Avoid infinite loops by checking the map.  

This is identical to:
- Copying linked structures  
- Rebuilding adjacency lists  
- Reconstructing computational graphs  


## 🪟 BFS Alternative

```python
from collections import deque

def cloneGraph_bfs(node):
    if not node:
        return None

    old_to_new = {node: Node(node.val)}
    queue = deque([node])

    while queue:
        curr = queue.popleft()

        for nei in curr.neighbors:
            if nei not in old_to_new:
                old_to_new[nei] = Node(nei.val)
                queue.append(nei)

            old_to_new[curr].neighbors.append(old_to_new[nei])

    return old_to_new[node]
```

## ⏱️ Time & Space Complexity

### Time Complexity: **O(V + E)**  
We visit every node and every edge once.

### Space Complexity: **O(V)**  
For the hash map and recursion/queue.

Efficient for large graphs.


## ⚠️ Edge Cases

- Empty graph → return None  
- Single node with no neighbors  
- Cycles (1 ↔ 2)  
- Self-loops (node points to itself)  
- Large connected components
