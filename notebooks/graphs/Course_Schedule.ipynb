# 207. Course Schedule  
**Category:** Graphs / Topological Sort / Cycle Detection  
**Difficulty:** Medium  

You are given `numCourses` and a list of prerequisite pairs `prerequisites`,  
where each pair `[a, b]` means: to take course `a`, you must first take course `b`.

Return **True** if you can finish all courses, otherwise **False**.

Example:  
Input: numCourses = 2, prerequisites = [[1,0]]  
Output: True

Input: numCourses = 2, prerequisites = [[1,0],[0,1]]  
Output: False  (cycle)


## 🎯 Intuition & ML Relevance

This problem is about **dependency resolution**, which appears everywhere in ML systems:

- DAGs in computational graphs  
- Layer dependencies in neural networks  
- Pipeline ordering (ETL → preprocessing → training → evaluation)  
- Scheduling tasks with prerequisites  

The course graph is a **directed graph**:
- Nodes = courses  
- Edges = prerequisites  

To finish all courses:
- The graph must be a **DAG** (Directed Acyclic Graph).  
- If there is a cycle → impossible to complete.

This problem teaches:
- Topological sorting  
- Cycle detection in directed graphs  
- Graph construction using adjacency lists  


## 🧩 Code Implementation (BFS Topological Sort)

```python
from collections import defaultdict, deque

def canFinish(numCourses, prerequisites):
    graph = defaultdict(list)
    indegree = [0] * numCourses

    # Build graph
    for course, prereq in prerequisites:
        graph[prereq].append(course)
        indegree[course] += 1

    # Start with courses that have no prerequisites
    queue = deque([i for i in range(numCourses) if indegree[i] == 0])
    taken = 0

    while queue:
        curr = queue.popleft()
        taken += 1

        for neighbor in graph[curr]:
            indegree[neighbor] -= 1
            if indegree[neighbor] == 0:
                queue.append(neighbor)

    return taken == numCourses

# Example
canFinish(2, [[1,0]])
```

## 🧠 Graph Logic (Topological Sort)

We use **Kahn’s Algorithm**:

1. Compute indegree for each course.  
2. Add all courses with indegree 0 to a queue.  
3. Repeatedly remove a course from the queue:  
   - "Take" the course  
   - Reduce indegree of its neighbors  
   - If a neighbor reaches indegree 0 → add to queue  
4. If we take all courses → no cycle → return True  
5. If some courses remain → cycle → return False  

This is identical to:
- Scheduling tasks with dependencies  
- Ordering ML pipeline stages  
- Ensuring DAG validity in computational graphs  


## 🪟 DFS Cycle Detection Alternative

```python
def canFinish_dfs(numCourses, prerequisites):
    graph = defaultdict(list)
    for course, prereq in prerequisites:
        graph[prereq].append(course)

    visited = [0] * numCourses  # 0 = unvisited, 1 = visiting, 2 = visited

    def dfs(course):
        if visited[course] == 1:
            return False  # cycle
        if visited[course] == 2:
            return True

        visited[course] = 1
        for nei in graph[course]:
            if not dfs(nei):
                return False

        visited[course] = 2
        return True

    for c in range(numCourses):
        if not dfs(c):
            return False

    return True
```

## ⏱️ Time & Space Complexity

### BFS Topological Sort
- Time: **O(V + E)**  
- Space: **O(V + E)**  

### DFS Cycle Detection
- Time: **O(V + E)**  
- Space: **O(V + E)**  

Efficient for large dependency graphs.


## ⚠️ Edge Cases

- No prerequisites → always True  
- Single course  
- Cycle of length 2 (A → B → A)  
- Long chain of dependencies  
- Disconnected components
