# 417. Pacific Atlantic Water Flow  
**Category:** Graphs / DFS / BFS / Matrix Reachability  
**Difficulty:** Medium  

You are given an `m x n` matrix of heights.  
Water can flow from a cell to its **4 neighbors** (up, down, left, right) *only if the neighbor’s height is less than or equal to the current cell*.

The Pacific Ocean touches the **top and left** edges.  
The Atlantic Ocean touches the **bottom and right** edges.

Return all coordinates where water can flow to **both** oceans.

Example:
Input:
[
  [1,2,2,3,5],
  [3,2,3,4,4],
  [2,4,5,3,1],
  [6,7,1,4,5],
  [5,1,1,2,4]
]

Output:
[[0,4],[1,3],[1,4],[2,2],[3,0],[3,1],[4,0]]


## 🎯 Intuition & ML Relevance

This problem is about **reverse graph traversal** and **reachability**, which appear in:

- Graph neural networks  
- Flow-based models  
- Influence propagation  
- Region-based reasoning  
- Constraint-based traversal  

Key insight:
- Instead of simulating water flowing *downhill*, we reverse the logic:
  - Start from the oceans  
  - Move *uphill* (to equal or higher heights)  
  - Mark all reachable cells  

Cells reachable from:
- Pacific → pacific_reachable  
- Atlantic → atlantic_reachable  

Intersection = cells that can reach **both** oceans.

This mirrors ML concepts such as:
- Reverse message passing  
- Bidirectional reachability  
- Region labeling  
- Multi-source BFS/DFS  


## 🧩 Code Implementation (DFS)

```python
def pacificAtlantic(heights):
    if not heights:
        return []

    rows, cols = len(heights), len(heights[0])

    pacific = set()
    atlantic = set()

    def dfs(r, c, visited, prev_height):
        if (
            r < 0 or c < 0 or
            r >= rows or c >= cols or
            (r, c) in visited or
            heights[r][c] < prev_height
        ):
            return

        visited.add((r, c))

        dfs(r + 1, c, visited, heights[r][c])
        dfs(r - 1, c, visited, heights[r][c])
        dfs(r, c + 1, visited, heights[r][c])
        dfs(r, c - 1, visited, heights[r][c])

    # Pacific: top row + left column
    for c in range(cols):
        dfs(0, c, pacific, heights[0][c])
    for r in range(rows):
        dfs(r, 0, pacific, heights[r][0])

    # Atlantic: bottom row + right column
    for c in range(cols):
        dfs(rows - 1, c, atlantic, heights[rows - 1][c])
    for r in range(rows):
        dfs(r, cols - 1, atlantic, heights[r][cols - 1])

    # Intersection
    return list(pacific & atlantic)
```

## 🧠 Graph Traversal Logic

Instead of flowing water from every cell → ocean (expensive), we reverse the direction:

1. Start DFS/BFS from the Pacific borders.  
2. Move only to neighbors with **equal or greater height**.  
3. Mark all reachable cells.  
4. Repeat for Atlantic borders.  
5. The intersection of reachable sets = cells that can reach both oceans.

Why reverse?
- Water flows downhill → many paths.
- Reverse flow (uphill) → only valid paths.
- Much more efficient.


## 🪟 BFS Alternative

```python
from collections import deque

def pacificAtlantic_bfs(heights):
    if not heights:
        return []

    rows, cols = len(heights), len(heights[0])

    def bfs(starts):
        visited = set(starts)
        queue = deque(starts)

        while queue:
            r, c = queue.popleft()

            for dr, dc in [(1,0),(-1,0),(0,1),(0,-1)]:
                nr, nc = r + dr, c + dc

                if (
                    0 <= nr < rows and
                    0 <= nc < cols and
                    (nr, nc) not in visited and
                    heights[nr][nc] >= heights[r][c]
                ):
                    visited.add((nr, nc))
                    queue.append((nr, nc))

        return visited

    pacific_starts = [(0, c) for c in range(cols)] + [(r, 0) for r in range(rows)]
    atlantic_starts = [(rows - 1, c) for c in range(cols)] + [(r, cols - 1) for r in range(rows)]

    pacific = bfs(pacific_starts)
    atlantic = bfs(atlantic_starts)

    return list(pacific & atlantic)
```

## ⏱️ Time & Space Complexity

### Time Complexity: **O(m · n)**
Each cell is visited at most twice (once for each ocean).

### Space Complexity: **O(m · n)**
For visited sets and recursion/queue.

Very efficient for large matrices.


## ⚠️ Edge Cases

- Empty matrix → []  
- Single row or column  
- All heights equal  
- Strictly increasing or decreasing terrain  
- Large grids (performance)
