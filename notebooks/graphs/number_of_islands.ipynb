# 200. Number of Islands  
**Category:** Graphs / Connectivity / DFS-BFS  
**Difficulty:** Medium  

Given a 2D grid of `'1'`s (land) and `'0'`s (water), count the number of **connected components** of land.  
Connectivity is vertical and horizontal (not diagonal).

Example:  
Input:
[
  ["1","1","0","0"],
  ["1","0","0","1"],
  ["0","0","1","1"]
]

Output: 3


## 🎯 Intuition & ML Relevance

This problem is a classic example of **graph connectivity**, which is heavily used in:

- Recommendation systems (user-item graph components)
- Community detection
- Content clustering
- Image segmentation
- Graph neural networks
- Pinterest’s discovery graph (pins, boards, users)

Understanding DFS/BFS traversal is essential for:
- Exploring connected components  
- Detecting clusters  
- Building graph-based ML features  


```python
def numIslands(grid):
    if not grid:
        return 0
    
    rows, cols = len(grid), len(grid[0])
    visited = set()
    islands = 0

    def dfs(r, c):
        # Out of bounds or water or already visited
        if (
            r < 0 or c < 0 or 
            r >= rows or c >= cols or 
            grid[r][c] == "0" or 
            (r, c) in visited
        ):
            return
        
        visited.add((r, c))
        
        # Explore neighbors
        dfs(r+1, c)
        dfs(r-1, c)
        dfs(r, c+1)
        dfs(r, c-1)

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1" and (r, c) not in visited:
                dfs(r, c)
                islands += 1
    
    return islands

# Example
grid = [
  ["1","1","0","0"],
  ["1","0","0","1"],
  ["0","0","1","1"]
]
numIslands(grid)
```

## 🌊 DFS Approach Explanation

We treat the grid as a graph:
- Each cell is a node
- Edges exist between adjacent land cells (up, down, left, right)

Algorithm:
1. Loop through every cell.
2. When we find unvisited land (`"1"`):
   - Start a DFS to mark the entire connected component.
   - Increment the island count.
3. Continue scanning the grid.

This ensures each island is counted exactly once.


## ⏱️ Time & Space Complexity

### Time Complexity: **O(rows × cols)**
We visit each cell at most once.

### Space Complexity:
- **O(rows × cols)** in the worst case (all land)
- DFS recursion stack + visited set

Efficient and scalable for large grids.


```python
from collections import deque

def numIslands_bfs(grid):
    if not grid:
        return 0
    
    rows, cols = len(grid), len(grid[0])
    visited = set()
    islands = 0

    for r in range(rows):
        for c in range(cols):
            if grid[r][c] == "1" and (r, c) not in visited:
                queue = deque([(r, c)])
                visited.add((r, c))

                while queue:
                    x, y = queue.popleft()
                    for dx, dy in [(1,0),(-1,0),(0,1),(0,-1)]:
                        nx, ny = x + dx, y + dy
                        if (
                            0 <= nx < rows and 
                            0 <= ny < cols and 
                            grid[nx][ny] == "1" and 
                            (nx, ny) not in visited
                        ):
                            visited.add((nx, ny))
                            queue.append((nx, ny))

                islands += 1

    return islands

# Example
numIslands_bfs(grid)
```

## ⚠️ Edge Cases

- Empty grid → return 0  
- All water → return 0  
- All land → return 1  
- Islands touching diagonally should NOT connect  
- Large grids (performance test)
