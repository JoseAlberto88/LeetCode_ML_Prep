# 451. Sort Characters by Frequency  
**Category:** Ranking / Frequency Analysis / Heaps / Strings  
**Difficulty:** Medium  

Given a string `s`, sort it in **descending order of character frequency**.

Example:  
Input: "tree"  
Output: "eert"  
Explanation: 'e' appears twice, 'r' and 't' once → "eert" or "eetr"


## 🎯 Intuition & ML Relevance

This problem is about ranking items (characters) by **frequency**, which is a core idea in:

- Token frequency analysis  
- Vocabulary pruning  
- Feature ranking  
- Histogram-based modeling  

Why this matters:

- Each character is represented by a **frequency embedding**.
- Sorting by frequency mirrors:
  - TF-IDF weighting  
  - Token importance scoring  
  - Feature selection in NLP  

This problem teaches:
- How to build frequency maps  
- How to sort by frequency  
- How to use heaps or bucket sort for ranking  


## 🧩 Code Implementation (Heap)

```python
from collections import Counter
import heapq

def frequencySort(s):
    freq = Counter(s)

    # Max-heap using negative frequencies
    heap = [(-count, char) for char, count in freq.items()]
    heapq.heapify(heap)

    result = []
    while heap:
        count, char = heapq.heappop(heap)
        result.append(char * (-count))

    return "".join(result)

# Example
frequencySort("tree")
```

## 🧠 Ranking Logic

Steps:
1. Count how many times each character appears.  
2. Use a heap to always extract the **highest-frequency** character.  
3. Append each character repeated by its frequency.  

This mirrors:
- Ranking tokens by importance  
- Sorting features by weight  
- Building histograms for NLP  


## 🪟 Bucket Sort Optimization (O(n))

```python
def frequencySort_bucket(s):
    freq = Counter(s)

    # Buckets: index = frequency
    buckets = [[] for _ in range(len(s) + 1)]

    for char, count in freq.items():
        buckets[count].append(char)

    result = []
    for count in range(len(buckets) - 1, 0, -1):
        for char in buckets[count]:
            result.append(char * count)

    return "".join(result)
```

## ⏱️ Time & Space Complexity

### Heap Method
- Time: **O(n log n)**  
- Space: **O(n)**  

### Bucket Sort Method
- Time: **O(n)**  
- Space: **O(n)**  

Bucket sort is optimal when frequencies are bounded by string length.


## ⚠️ Edge Cases

- Empty string → ""  
- All characters identical  
- All characters unique  
- Very long strings (performance)
