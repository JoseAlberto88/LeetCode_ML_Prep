# 347. Top K Frequent Elements  
**Category:** Ranking / Frequency Analysis / Heaps  
**Difficulty:** Medium  

Given an integer array `nums` and an integer `k`, return the **k most frequent elements**.

Example:  
Input: nums = [1,1,1,2,2,3], k = 2  
Output: [1, 2]


## 🎯 Intuition & ML Relevance

This problem is about ranking items by **frequency**, which is central to many ML workflows:

- Feature selection  
- Token frequency analysis  
- Histogram-based ranking  
- Identifying dominant patterns in data  

Why this matters:

- We build a **frequency embedding** for each number.
- We then extract the **top-k** most frequent elements.
- This mirrors ranking tasks in ML such as:
  - Selecting top features  
  - Identifying dominant clusters  
  - Prioritizing high-signal inputs  

This problem teaches:
- Efficient counting  
- Heap-based top-k retrieval  
- Sorting by frequency  


## 🧩 Code Implementation (Heap)

```python
from collections import Counter
import heapq

def topKFrequent(nums, k):
    freq = Counter(nums)

    # Build a max-heap using negative frequencies
    heap = [(-count, num) for num, count in freq.items()]
    heapq.heapify(heap)

    result = []
    for _ in range(k):
        result.append(heapq.heappop(heap)[1])

    return result

# Example
topKFrequent([1,1,1,2,2,3], 2)
```

## 🧠 Ranking Logic

Steps:
1. Count frequencies of each number.  
2. Use a heap to efficiently extract the top-k elements.  
3. Higher frequency → higher priority.  

This is identical to:
- Ranking features by importance  
- Selecting top histogram bins  
- Prioritizing high-signal data points  


## 🪟 Bucket Sort Optimization (O(n))

```python
def topKFrequent_bucket(nums, k):
    freq = Counter(nums)

    # Create buckets: index = frequency
    buckets = [[] for _ in range(len(nums) + 1)]

    for num, count in freq.items():
        buckets[count].append(num)

    result = []
    # Traverse buckets from high frequency to low
    for count in range(len(buckets) - 1, 0, -1):
        for num in buckets[count]:
            result.append(num)
            if len(result) == k:
                return result
```

## ⏱️ Time & Space Complexity

### Heap Method
- Time: **O(n log n)**  
- Space: **O(n)**  

### Bucket Sort Method
- Time: **O(n)**  
- Space: **O(n)**  

Bucket sort is optimal when frequencies are bounded by `n`.


## ⚠️ Edge Cases

- All elements identical  
- All elements unique  
- Large arrays (performance)
