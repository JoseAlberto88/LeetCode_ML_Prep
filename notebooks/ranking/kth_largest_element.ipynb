## 📝 Problem Summary

Given an integer array `nums` and an integer `k`, return the **k-th largest element** in the array.

Important:
- The k-th largest element is based on **sorted order**, not distinctness.
- Example:  
  Input: nums = [3,2,1,5,6,4], k = 2  
  Output: 5


## 🎯 Intuition & ML Relevance

This problem is directly related to **ranking**, which is fundamental in:

- Recommendation systems (e.g., top-K items)
- Search engines (e.g., top-K results)
- Feature selection (e.g., top-K important features)
- Retrieval systems (e.g., nearest neighbors)

Efficiently selecting the k-th largest element is more scalable than fully sorting the array.


```python
import heapq

def kthLargest(nums, k):
    # Python's heapq is a min-heap, so we keep a heap of size k
    min_heap = []
    
    for num in nums:
        heapq.heappush(min_heap, num)
        
        # Keep only k largest elements in the heap
        if len(min_heap) > k:
            heapq.heappop(min_heap)
    
    return min_heap[0]

# Example
nums = [3,2,1,5,6,4]
k = 2
kthLargest(nums, k)
```

## ⏱️ Time & Space Complexity

### Heap Approach
- **Time:** O(n log k)  
  We push/pop at most n times, each heap operation costs log k.
- **Space:** O(k)  
  We store only the k largest elements.

### Why this is efficient:
- Sorting would cost O(n log n)
- Heap reduces unnecessary work
- Scales better for large datasets (important in ML pipelines)


```python
import random

def quickselect(nums, k):
    k = len(nums) - k  # convert to k-th smallest index
    
    def select(left, right):
        pivot = nums[random.randint(left, right)]
        l, r = left, right
        
        while l <= r:
            while nums[l] < pivot:
                l += 1
            while nums[r] > pivot:
                r -= 1
            if l <= r:
                nums[l], nums[r] = nums[r], nums[l]
                l += 1
                r -= 1
        
        if k <= r:
            return select(left, r)
        if k >= l:
            return select(l, right)
        return nums[k]
    
    return select(0, len(nums) - 1)

# Example
quickselect([3,2,1,5,6,4], 2)
```

## ⚠️ Edge Cases

- k = 1 → return the maximum element
- k = len(nums) → return the minimum element
- nums contains duplicates
- nums contains negative numbers
- nums is already sorted (ascending or descending)
