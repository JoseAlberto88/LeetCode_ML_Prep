# 102. Binary Tree Level Order Traversal  
**Category:** Trees / BFS Traversal  
**Difficulty:** Medium  

Given the root of a binary tree, return the **level order traversal** of its nodes' values.  
(From left to right, level by level.)

### Example

Input:
    3
   / \
  9  20
     / \
    15  7

Output:
[
  [3],
  [9,20],
  [15,7]
]


# 🎯 Intuition & ML Relevance

This problem is the foundation of **Breadth‑First Search (BFS)** on trees.

It demonstrates:

- Level‑by‑level processing  
- Queue‑based traversal  
- Hierarchical reasoning  
- Layered feature extraction  

ML relevance:

- Decision tree layer extraction  
- Hierarchical clustering

- Neural network layer interpretation  
- Tree‑structured data traversal  

The key idea:

> Use a queue to process nodes level by level.


# 🟦 BFS Solution (Queue)

```python
from collections import deque

def levelOrder(root):
    if not root:
        return []

    result = []
    queue = deque([root])

    while queue:
        level = []
        size = len(queue)

        for _ in range(size):
            node = queue.popleft()
            level.append(node.val)

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

        result.append(level)

    return result
```

# ⚙️ BFS Explanation

We use a queue to process nodes in FIFO order.

For each level:

1. Determine the number of nodes at this level (`size`)  
2. Pop exactly that many nodes from the queue  
3. Add their children to the queue  
4. Append the collected values to the result  

This ensures we process the tree **level by level**, left to right.


# 🟦 DFS Version (Alternative)

```python
def levelOrder_dfs(root):
    result = []

    def dfs(node, level):
        if not node:
            return

        if len(result) == level:
            result.append([])

        result[level].append(node.val)

        dfs(node.left, level + 1)
        dfs(node.right, level + 1)

    dfs(root, 0)
    return result
```

# ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
We visit each node once.

### Space Complexity:  
- **O(n)** for the queue (worst case: last level full)  
- **O(h)** recursion stack for DFS version  


# ⚠️ Edge Cases

- Empty tree → return []  
- Single node → [[value]]  
- Unbalanced trees → BFS still works  
- Deep trees → DFS recursion depth may be large
