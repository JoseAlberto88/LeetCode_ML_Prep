# 112. Path Sum  
**Category:** Trees / DFS / Root‑to‑Leaf Reasoning  
**Difficulty:** Easy  

Given the root of a binary tree and an integer `targetSum`,  
return **True** if the tree has a **root‑to‑leaf path** such that  
adding up all the values along the path equals `targetSum`.

A **leaf** is a node with no children.

### Example

Input:
    5
   / \
  4   8
 /   / \
11  13  4
/  \      \
7    2      1

targetSum = 22

Output: True  
Explanation: 5 → 4 → 11 → 2 = 22


# 🎯 Intuition & ML Relevance

This problem is a classic example of **recursive depth‑first search**  
with **state accumulation**.

It demonstrates:

- Root‑to‑leaf reasoning  
- Recursive state passing  
- Decision path evaluation  
- Tree traversal fundamentals  

ML relevance:

- Decision tree path evaluation  
- Computing cumulative values along hierarchical structures  
- Recursive feature extraction  
- Reasoning over structured data  

The key idea:

> At each node, subtract its value from the target  
> and recursively check if any child can complete the path.


# 🟦 Recursive DFS Solution

```python
def hasPathSum(root, targetSum):
    if not root:
        return False

    # If it's a leaf, check if the remaining sum matches the node value
    if not root.left and not root.right:
        return targetSum == root.val

    # Recurse on children with reduced target
    new_target = targetSum - root.val

    return (
        hasPathSum(root.left, new_target) or
        hasPathSum(root.right, new_target)
    )
```

# ⚙️ DFS Explanation

At each node:

1. Subtract the node's value from `targetSum`  
2. If it's a leaf, check if the remaining sum is zero  
3. Otherwise, recursively check left and right children  

This is a classic DFS pattern with state accumulation.

The recursion naturally handles:

- Leaf detection  
- Null children  
- Arbitrary tree shapes  


# 🟦 Iterative DFS Version (Stack)

```python
def hasPathSum_iterative(root, targetSum):
    if not root:
        return False

    stack = [(root, targetSum - root.val)]

    while stack:
        node, remaining = stack.pop()

        # Leaf check
        if not node.left and not node.right and remaining == 0:
            return True

        if node.left:
            stack.append((node.left, remaining - node.left.val))
        if node.right:
            stack.append((node.right, remaining - node.right.val))

    return False
```

# ⏱️ Time & Space Complexity

### Time Complexity: **O(n)**  
We visit each node once.

### Space Complexity:  
- **O(h)** recursion stack (h = tree height)  
- **O(n)** worst case for skewed trees  


# ⚠️ Edge Cases

- Empty tree → False  
- Single node → True only if node.val == targetSum  
- Negative values allowed  
- Unbalanced trees still work
